```python
from perplexity import Perplexity
from sqlalchemy.util import await_only
from dotenv import load_dotenv
import os
load_dotenv()

client = Perplexity(api_key=os.getenv("PERPLEXITY_API_KEY"))

search = client.search.create(
    query="Explain concisely what is GRPO?",
    max_results=10,
    max_tokens=25000,
    max_tokens_per_page=2048,
    country="VN",
    search_language_filter=["en"],
    search_recency_filter="week",
)

for result in search.results:
    print(f"{result.title}: {result.url}")
```

## Alternative way to call api

```python
import requests

api_key = os.getenv("PERPLEXITY_API_KEY")
url = os.getenv("PERPLEXITY_URL")

payload = {
    "query": "How can I combine fine-tuning and RAG to improve model performance?",
    "max_results": 3,
    "max_tokens": 25000,
    "max_tokens_per_page": 2048,
    "country": "VN",
    "search_language_filter": ["en"],
    "search_recency_filter": "week"
}

headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}

resp = requests.post(url, json=payload, headers=headers)
print(resp.status_code)
print(resp.json())
```
